```python
import pandas as pd
from matplotlib import pyplot as plt
data = pd.read_csv('fill_data.csv')
ages = data['Age']
dev_salaries = data['All_Devs']
py_salaries = data['Python']
js_salaries = data['JavaScript']

plt.plot(ages,dev_salaries,color='blue',linestyle='--',label='All devs',linewidth=2)
plt.plot(ages,py_salaries,color='red',linestyle='-',label='Python',linewidth=3)
#plt.plot(ages,js_salaries,color='yellow',linestyle='-',label='Javascript',linewidth=2)

overall_median = 57287
#Lower uppper all fill
#plt.fill_between(ages,py_salaries,overall_median,color='green',alpha=0.25)

#Only upper than overall_median
#plt.fill_between(ages,py_salaries,overall_median,where=(py_salaries > overall_median),interpolate=True,
#              alpha=0.25,color='blue')

#Compare to dev_salaries,more than dev_salaries
plt.fill_between(ages,py_salaries,dev_salaries,where=(py_salaries > dev_salaries),interpolate=True,
                 alpha=0.25,color='blue',label='python_Above_all_dev')


    
#Only less than overall_median

#plt.fill_between(ages,py_salaries,overall_median,where=(py_salaries <= overall_median),interpolate=True,
#                alpha=0.75,color='orange')

#Compare to dev_salaries,less than dev_salaries
plt.fill_between(ages,py_salaries,dev_salaries,where=(py_salaries <= dev_salaries),interpolate=True,
                 alpha=0.75,color='orange',label='python_below_all_dev')



plt.legend(loc='upper left')
plt.title('Developer income')
plt.xlabel('Age')
plt.ylabel('Income')
```

```python
from matplotlib import pyplot as plt
import pandas as pd
from collections import Counter

#print(plt.style.available)
plt.style.use("ggplot")

data_csv=pd.read_csv('./data.csv')
data=data_csv['LanguagesWorkedWith']
counter=Counter()
for da in data:
    counter.update(da.split(';'))
print(counter.most_common(5))
slices=[]
labels=[]
for label,slice in (counter.most_common(5)):
    slices.append(slice)
    labels.append(label)

explode = [0,0,0,0.05,0]
#colors = ['blue','red','yellow','green']

#plt.pie(slices,labels=labels,colors=colors,wedgeprops={'edgecolor':'white'})
plt.pie(slices,labels=labels,explode=explode,shadow=True,wedgeprops={'edgecolor':'white'},
        startangle=45,autopct='%1.1f%%')
plt.title("Pie chart ex1")
plt.tight_layout()
plt.show()
```
